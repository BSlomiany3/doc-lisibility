### **Fonctions d’activation de base**

| Nom       | PyTorch                | Fonctionnelle                       | Description |
|-----------|------------------------|-------------------------------------|-------------|
| ReLU      | `nn.ReLU()`            | `F.relu(x)`                         | Met à zéro les valeurs négatives |
| Sigmoid   | `nn.Sigmoid()`         | `F.sigmoid(x)`                      | Écrase entre 0 et 1 |
| Tanh      | `nn.Tanh()`            | `F.tanh(x)`                         | Écrase entre -1 et 1 |

```python
# imports
import matplotlib.pyplot as plt

import torch
from torch import nn
```

#### **ReLU**
---

```python
x = torch.linspace(-15, 15, 1000, requires_grad=True)
relu = nn.ReLU() # torch.max(torch.tensor(0), x)

y = relu(x)
dy = torch.where(x > 0, torch.ones_like(x), torch.zeros_like(x))

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# ReLU
axes[0].plot(x.detach().numpy(), y.detach().numpy(), color='blue')
axes[0].set_title("Fonction ReLU")
axes[0].set_xlabel("x")
axes[0].set_ylabel("ReLU(x)")
axes[0].grid(True)

# ReLU'
axes[1].plot(x.detach().numpy(), dy.detach().numpy(), color='red')
axes[1].set_title("Dérivée de ReLU")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f'(x)")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

#### **Sigmoide**
---

```python
x = torch.linspace(-15, 15, 1000)
sigmoid = nn.Sigmoid() # (1 / (1 + torch.exp(-x)))

y = sigmoid(x)
dy = y * (1 - y)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Sigmoïde
axes[0].plot(x.numpy(), y.numpy(), color='blue')
axes[0].set_title("Sigmoïde")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].grid(True)

# Dérivée
axes[1].plot(x.numpy(), dy.numpy(), color='red')
axes[1].set_title("Dérivée de la sigmoïde")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f'(x)")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

#### **TanH**
---

```python
x = torch.linspace(-15, 15, 1000)
tanh = nn.Tanh() # ((torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x)))

y = tanh(x)
dy = 1 - y**2

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# tanh
axes[0].plot(x.numpy(), y.numpy(), color='blue')
axes[0].set_title("Fonction tanh")
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")
axes[0].grid(True)

# dérivée
axes[1].plot(x.numpy(), dy.numpy(), color='red')
axes[1].set_title("Dérivée")
axes[1].set_xlabel("x")
axes[1].set_ylabel("f'(x)")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
